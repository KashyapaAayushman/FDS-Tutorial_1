```python
from numpy import random
x = random.rand(100)
print(x)
```

```python
from numpy import random
x = random.choice([4,5,2,8], p=[0.1,0.3,0.4,0.2], size=(100))
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.normal(size=6000), hist=True)
plt.show()
```

```python
from numpy import random
x = random.binomial(n=10, p=0.6, size=10)
print(x)
```

```python
from numpy import random
x = random.exponential(n=10, p=0.6, size=10)
print(x)
```
